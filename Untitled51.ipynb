```python
#model selection
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 41)
```

```python
#logistic regression
#model training
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,mean_squared_error,root_mean_squared_error
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
a = accuracy_score(y_test,y_pred)
b = confusion_matrix(y_test,y_pred)
c = mean_squared_error(y_test,y_pred)
d = root_mean_squared_error(y_test,y_pred)
print(a)
print(b)
print(c)
print(d)
```

```python
#descision tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
a = accuracy_score(y_test,y_pred)
b = confusion_matrix(y_test,y_pred)
c = mean_squared_error(y_test,y_pred)
d = root_mean_squared_error(y_test,y_pred)
print(a)
print(b)
print(c)
print(d)
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
a = accuracy_score(y_test,y_pred)
b = confusion_matrix(y_test,y_pred)
c = mean_squared_error(y_test,y_pred)
d = root_mean_squared_error(y_test,y_pred)
print(a)
print(b)
print(c)
print(d)
```

```python
#finding outliers in data
plt.figure(figsize = (6,4))
sns.boxplot(x = df["Age"])
plt.title("boxplot")
plt.show()
```

```python
#in this there is no outliers but we try ti apply iqr to remove outliers
Q1 = df["Age"].quantile(0.25)
Q3 = df["Age"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print("Outlier range: <", lower_bound, "or >", upper_bound)
```

```python
df_no_outliers = df[
    (df["Age"] >= lower_bound) & (df["Age"] <= upper_bound)
]

print("After removing outliers:", df_no_outliers.shape)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df_no_outliers["Age"])
plt.title("Boxplot")
plt.show()
```
